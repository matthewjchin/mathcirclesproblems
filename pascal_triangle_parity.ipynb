```python
from scipy import linalg as sl
import numpy as np
```

```python
rows = 750
```

```python
result = sl.pascal(rows, kind="lower")
```

```python
print(result)
```

```python
i = 0
for row in result:
    #print(row % 2 == 1)
    print("row: ", i, "     odds: ", np.sum(row % 2 == 1))
    i += 1
```
